```python
# Define the file paths
file_path_importaciones = '/content/drive/MyDrive/Colab Notebooks/Data/Proyecto/dataset filtrado importaciones mayo 2024.xlsx'
file_path_arancelarias = '/content/drive/MyDrive/Colab Notebooks/Data/Proyecto/Dataset partidas arancelarias 3.xlsx'

# Read the sheets individually
importaciones_df = pd.read_excel(file_path_importaciones, sheet_name='Sheet1')
arancelarias_df = pd.read_excel(file_path_arancelarias, sheet_name='Table 1')

# Display the first few rows of each dataframe
print('Importaciones DataFrame:')
print(importaciones_df.head())
print('\
')

print('Arancelarias DataFrame:')
print(arancelarias_df.head())
print('')
```

```python

# Merge the datasets
merged_df = pd.merge(importaciones_df, arancelarias_df[['SUBPARTIDA_ARANCELARIA', 'Numero Seccion']],
                     on='SUBPARTIDA_ARANCELARIA', how='left')

# Count the number of importations for each section
section_counts = merged_df['Numero Seccion'].value_counts().sort_index()

# Create a bar plot
plt.figure(figsize=(15, 8))
sns.barplot(x=section_counts.index, y=section_counts.values)
plt.title('Cantidad de Importaciones por Número de Sección')
plt.xlabel('Número de Sección')
plt.ylabel('Número de Importaciones')
plt.xticks(rotation=0)
plt.show()

# Print the counts
print("Cantidad de Importaciones por cada Sección:")
print(section_counts)

# Calculate the percentage of importations for each section
total_importations = section_counts.sum()
section_percentages = (section_counts / total_importations * 100).round(2)

# Print the percentages
print("\
Porcentaje de Importaciones por cada Sección:")
print(section_percentages)
```

```python
# Definir las categorias de interes
categories_of_interest = [
    'Animales vivos y productos del reino animal',
    'Instrumentos y aparatos de óptica, fotografía',
    'Mercancía y productos diversos',
    'Materias textiles y sus manufacturas',
    'Material de transporte'
                         ]


# Merge the datasets
merged_df = pd.merge(importaciones_df, arancelarias_df[['SUBPARTIDA_ARANCELARIA', 'Designacion Seccion']],
                     on='SUBPARTIDA_ARANCELARIA', how='left')

# Filter for the categories of interest
filtered_df = merged_df[merged_df['Designacion Seccion'].isin(categories_of_interest)]

# Group by 'Designacion Seccion' and count the number of importations
category_counts = filtered_df['Designacion Seccion'].value_counts()

# Create a DataFrame for the results
results_df = pd.DataFrame({
    'Categoria': category_counts.index,
    'Volumen de Importaciones': category_counts.values
})

# Sort the results by volume in descending order
results_df = results_df.sort_values('Volumen de Importaciones', ascending=False)

# Display the table
print("Tabla de Volumen de Importaciones por Categoria:")
print(results_df.to_string(index=False))

# Create a bar plot
plt.figure(figsize=(8, 3))
sns.barplot(x='Volumen de Importaciones', y='Categoria', data=results_df)
plt.title('Volumen de Importaciones por Categoria')
plt.xlabel('Volumen de Importaciones')
plt.ylabel('Categoria')
plt.tight_layout()
plt.show()

# Calculate percentages
total_imports = results_df['Volumen de Importaciones'].sum()
results_df['Porcentaje'] = (results_df['Volumen de Importaciones'] / total_imports * 100).round(2)

print("\
Tabla de Volumen de Importaciones por Categoria (con porcentajes):")
print(results_df.to_string(index=False))
```
